```python
import scraping3 as sc
```

```python
url_pattern = 'https://store.ubi.com/fr/all-games?prefn1=productTypeRefinementString&sz=16&format=page-element&prefv1=games&start=%s'
```

```python
pages_to_scrape = 15
```

```python
spider= sc.MySpider(url_pattern, pages_to_scrape, scraper = sc.soup_pretty)
```

```python
def kickstarter(spider=spider):
    from multiprocessing import Pool
    list_of_links_to_scrape = [spider.url_pattern%i for i in range(1,spider.pages_to_scrape*16)]
    pool=Pool(5)
    res=pool.map(spider.scrape_url, list_of_links_to_scrape)
    pool.close()
    pool.join()
    return res
```

```python
[spider.url_pattern%i for i in range(1,spider.pages_to_scrape*16)]
```

```python
import pandas as pd
df=pd.DataFrame(kickstarter()[0])
df.sample(15)
```
